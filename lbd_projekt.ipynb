```python
def parse(series, key):
    tmp = series.copy(deep=True)
    tmp = tmp.map(lambda x: json.loads(x))
    tmp = tmp.map(lambda x: list(map(lambda y: y[key], x)))
    return tmp
```

```python
foo = movies.copy()

columns = ['id', 'original_title']
parsedColumns = []
toParse = [ 
    { 'col': 'genres', 'key': 'name' },
    { 'col': 'keywords', 'key': 'name'},
    { 'col': 'production_countries', 'key': 'name'},
    { 'col': 'production_companies', 'key': 'name'}
]

for x in toParse:
    foo[x['col']] = parse(foo[x['col']], x['key'])
    parsedColumns.append(x['col'])
        
foo = foo.filter(columns + parsedColumns)
display(foo)
```

```python
def splitColumns(data, columns):
    tmp = data.copy(deep=True)
    for index, row in tmp.iterrows():
        for col in columns:
            for x in row[col]:
                tmp.at[index, x] = 1
    tmp = tmp.drop(columns=columns)
    tmp = tmp.fillna(0)
    return tmp


def buildMovieRatings(data, ratings):
    titles = ratings.keys()
    ratedMovies = data.copy()
    ratedMovies = ratedMovies[ratedMovies['original_title'].isin(titles)]
    ratedMovies = ratedMovies.filter(['id', 'original_title'])
    ratedMovies['rating'] = ratedMovies.apply(lambda x: ratings[x['original_title']], axis=1)
    ratedMovies.reset_index(drop=True, inplace=True)
    return ratedMovies

def alterRatings(data, ratings):
    p = profile.copy(deep=True)
    for key in ratings:
        p[key] =  ratings[key] if p[key] == 0 else p[key]
    return p

def buildProfile(data, movieRatings):
    ratedMovies = data.copy()
    ratedMovies = ratedMovies[ratedMovies.id.isin(movieRatings.id)]

    ratedMovies.reset_index(drop=True, inplace=True)
    ratedMovies.drop(columns=columns, axis=1,inplace=True)
    
    profile = ratedMovies.T.dot(movieRatings.rating)
    return profile

def getDataWithoutRatedMovies(data, movieRatings):
    return data[~data.id.isin(movieRatings.id)]

def getRecom(data, profile, n):
    tmp = data.copy()
    tmp = tmp.drop(columns=columns)
    recom = (tmp.dot(profile)) / profile.sum()
    recom.sort_values(ascending=False, inplace=True)
    top5ind = recom.index[:n].tolist()
    recom_movies = movies.loc[top5ind,:]
    return recom_movies
```

```python
data = splitColumns(foo[0:100], parsedColumns)

movieRatings = {
    'Inception': 4.0,
    'Interstellar': 4.5, 
    'Pacific Rim': 3.0 
}

otherRatings = { 
    'Fantasy': 4,
    'Walt Disney Pictures': 50,
    'ocean': 50 
}

movieRatingsData = buildMovieRatings(data, movieRatings)

profile = buildProfile(data, movieRatingsData)
profile = alterRatings(profile, otherRatings)

data = getDataWithoutRatedMovies(data, movieRatingsData)
recom_movies = getRecom(data, profile, 20)

display(recom_movies['original_title'])
```
